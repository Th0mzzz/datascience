```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('brasil_copa2022.csv')
```

```python
def calcular_media_gols(df):
    media_gols = df['brasil_gols'].mean()
    return media_gols

media_gols = calcular_media_gols(df)
print(f'Média de gols por jogo: {media_gols:.2f}')

partidas_acima_media = 0
for partida in df.itertuples():
    if(partida.brasil_gols > media_gols):
        partidas_acima_media += 1
print(f'O Brasil marcou mais gols que a média em {partidas_acima_media} partidas.')
        
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["oponentes"], df["brasil_gols"], marker='s', label='Gols do Brasil')
plt.axhline(y=media_gols, color='r', linestyle='--', label=f'Média ({media_gols:.2f})')
plt.xticks(rotation=45)
plt.yticks(range(0, df["brasil_gols"].max() + 1))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df["oponentes"], df["oponentes_gols"])
plt.xticks(rotation=45)
plt.show()
```
